```
import pandas as pd

df = pd.read_csv("drive/My Drive/FERI/MAG/TRIR/Data/happiness_oecd.csv", sep=";", decimal=",")
df.shape
for col in df.columns:
  if col != "Region" and col != "Country":
    df[col] = pd.to_numeric(df[col])
df.dtypes
```

#### 1
###### Izri≈°i histogram za vrednost 'Happiness.Score'. Histogram naj ima le 20 stolpcev, brez linije.

```
import seaborn as sns
from matplotlib import pyplot as plt

sns.distplot(df['Happiness.Score'], bins=20, kde=False)
```

#### 2
###### Izri≈°ite scatterplot za vrednosti 'Happiness.Score' in 'GDP.per.Capita', s ƒçrto.

```
sns.lmplot(x='Happiness.Score', y='Economy..GDP.per.Capita.', data=df)
```

#### 3
###### Izri≈°ite scatterplot za vrednosti 'Happiness.Score' in 'Life.Expectancy', s ƒçrto.

```
sns.lmplot(x="Happiness.Score", y="Health..Life.Expectancy.", data=df)
```

#### 4
###### Izri≈°ite scatterplot za vrednosti 'Happiness.Score' in 'Freedom', brez ƒçrte.

```
sns.scatterplot(x="Happiness.Score", y="Freedom", data=df)
```

#### 5
###### Izri≈°ite scatterplot za vrednosti 'Happiness.Score' in 'Government.Corruption', brez ƒçrte, za vsako regijo posebej.

```
sns.scatterplot(x="Happiness.Score", y="Trust..Government.Corruption.", data=df, hue="Region")
```

#### 6
###### Izri≈°ite graf korelacij heatmap vseh ≈°tevilskih spremenljivk.

```
corr = df.corr()

sns.heatmap(corr)
```

#### 7
###### Izri≈°ite boxplot-e za vsako regijo za vrednost 'Happiness.Score'.

```
sns.boxplot(data=df, x="Happiness.Score", y="Region")
```

#### 8
###### Izri≈°ite barplot, kjer pre≈°tejete ≈°tevilo dr≈æav glede na regijo.

```
sns.countplot(data=df, x="Region")

plt.xticks(rotation=-80)
```

#### 9
###### Izri≈°ite barplot, kjer prika≈æete povpreƒçje 'GDP.per.Capita' glede na regijo.

```
import numpy as np

sns.barplot(data=df, x="Region", y="Economy..GDP.per.Capita.", estimator=np.mean)

plt.xticks(rotation=-80)
```

#### 10
###### Izri≈°ite barplot, kjer prika≈æete povpreƒçje 'Life.Expectancy' glede na regijo, spremenite barve v odtenke rdeƒçe.

```
sns.catplot(data=df, x="Region", y="Health..Life.Expectancy.", estimator=np.mean, palette='Reds_d', kind="bar", height=5.5, aspect=2.5)

plt.xticks(rotation=-80)
```

#### 11
###### Izri≈°ite barplot, kjer prika≈æete povpreƒçje 'Freedom' glede na regijo, spremenite barve v odtenke modre.


```
sns.catplot(data=df, kind="bar", x="Region", y="Freedom", palette="Blues_d", height=5.5, aspect=2.5, estimator=np.mean)

plt.xticks(rotation=-80)
```

#### 12
###### Izri≈°ite barplot, kjer prika≈æete povpreƒçje 'Government.Corruption' glede na regijo, spremenite barve v odtenke sive.

```
sns.catplot(data=df, kind="bar", estimator=np.mean, x="Region", y="Trust..Government.Corruption.", height=5.5, aspect=2.5, palette="Greys_d")

plt.xticks(rotation=-80)
```
