####1 
######Na voljo imate štiri datoteke, katere vse naložite v svoj DataFrame.

```
import pandas as pd

df_happ_2017 = pd.read_csv("drive/My Drive/FERI/MAG/TRIR/Data/happiness2017.csv", sep = ",", decimal = ".", index_col = "Country")
df_happ_2017.shape

df_regions = pd.read_csv("drive/My Drive/FERI/MAG/TRIR/Data/regions.csv", sep = "\t", names = ['Country', 'Region'], index_col = "Country")
df_regions.shape

col_names = ["Country", "Happiness Rank 2016", "Happiness Score 2016", "Happiness Rank 2015", "Happiness Score 2015"]
df_happ_2016_2015 = pd.read_excel("drive/My Drive/FERI/MAG/TRIR/Data/happiness_2016_2015.xlsx", index_col=0, sheet_name=0, header=None, names = col_names, skiprows=1)
df_happ_2016_2015.shape
df_happ_2016_2015.head
df_happ_2016_2015.dtypes

df_oecd = pd.read_excel("drive/My Drive/FERI/MAG/TRIR/Data/oecd.xlsx", index_col=0, sheet_name=0, skiprows=[27])
```

####2
######Vse DataFrame-e združite v en DataFrame (df).

```
df = pd.concat([df_happ_2017, df_regions, df_happ_2016_2015, df_oecd], axis = 1, sort = True)
df.index.name = "Country"
df.shape
df.tail()
```

####3
######Izpišite število vrstic, ki ima manjkajoče podatke, za vsak stolpec.

```
df.isna().sum()
```

####4
######Za manjkajočo regijo ročno (z specifičnim spreminjanjem ene celice) dodajte vse regije posamezno.

```
df["Region"].isnull()
bool_series = pd.isnull(df["Region"])
df[bool_series]["Region"]
```

```
df.loc[df.index == "Central African Republic", "Region"] = "Sub-Saharan Africa"
df.loc[df.index == "Djibouti", "Region"] = "Middle East and Northern Africa"
df.loc[df.index == "Hong Kong S.A.R., China", "Region"] = "Eastern Asia"
df.loc[df.index == "Korea", "Region"] = "Eastern Asia"
df.loc[df.index == "Lesotho", "Region"] = "Sub-Saharan Africa"
df.loc[df.index == "Mozambique", "Region"] = "Sub-Saharan Africa"
df.loc[df.index == "Oman", "Region"] = "Middle East and Northern Africa"
df.loc[df.index == "Slovak Republic", "Region"] = "Central and Eastern Europe"
df.loc[df.index == "Somaliland region", "Region"] = "Middle East and Northern Africa"
df.loc[df.index == "Swaziland", "Region"] = "Sub-Saharan Africa"
df.loc[df.index == "Taiwan Province of China", "Region"] = "Eastern Asia"
```

####5
######Za vse ostale manjkajoče vrednosti dodajte povprečje tistih stolpcev.

```
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

####9
######Dodajte sledeče stolpce: RankDiff2017 (Happiness Rank 2017 - Happiness Rank 2016) in RankDiff2016 (Happiness Rank 2016 - Happiness Rank 2015).

```
df["RankDiff2017"] = df["Happiness.Rank"] - df["Happiness Rank 2016"] 
df["RankDiff2016"] = df["Happiness Rank 2016"] - df["Happiness Rank 2015"]
df.head()
```
